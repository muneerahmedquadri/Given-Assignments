### Skill Test - 5
- After cleaning the Date Column, split it into two columns IF the date has a starting and ending range else don't split.
- For each date being split, calculate the average date and store it in a new column.

```python
for i in data:
    if i[6].startswith('-'):
        i[6] = i[6].replace('-', '')
        print(i[6])
```

```python
for i in data:
    if '-' in i[6]:
        value = i[6].split('-')
        F = int(value[0])
        S = int(value[1])
        add = round((F+S)/2)
        i[6] = add
        print(i[6])
```
