### Importing The Dataset

`Import your 'applestore' dataset with its header and data in different variables`

```python
from csv import reader
```

```python
dataset = list(reader(open('AppleStore.csv', encoding = 'utf-8')))
```

```python
header = dataset[0]
data = dataset[1:]
```

The dataset provides information related to the apps available on the apple store with its price, reviews, ratings, etc.

### Task 1 - Calculating Average User Rating

`For the whole dataset including all of its apps, calculate their average user rating`

```python
header.index('user_rating')
```

```python
avg_user_rating = []
for i in data:
    avg_user_rating.append(float(i[7]))
    avg = sum(avg_user_rating)/len(avg_user_rating)
print(avg)
```

### Task 3 - Calculate Average User Rating For Paid Music Or Paid Games

`Use the relative columns to calculate the average user rating for apps which are paid and are either in the 'Music' or the 'Games' category`

```python
avg_user_rating2 = []
for i in data:
    if float(i[4]) != 0 and (i[11] == 'Music' or i[11] == 'Games'):
        avg_user_rating2.append(float(i[7]))
        avg = sum(avg_user_rating2)/len(avg_user_rating2)
        
print(avg)
```

### Task 4 - Creating Price Category Using Conditional Statements

`Use the conditions below to create a new column named 'category' which specifies the type of product in relation to its price`

Example: All Applications with price greater than 20 and less than and equal to 50 should have Affordable in their category column

```python
header.append('Category')
```

```python
header
```

```python
for i in data:
    price = float(i[4])
    if price == 0:
        i.append('free')
    elif price > 0 and price <= 20:
        i.append('cheap')
    elif price > 20 and price <= 50:
        i.append('affordable')
    elif price > 50:
        i.append('expensive')
    else:
        i.append('price not given')
```
